As the project stated, there are two types of animals at the shop: Pet and Fish. They were made as two different tables because each of them need different accommodations. For example, fish need water and a tank and multiple fish can be in one tank (ex. goldfish). Also, pets like dogs and cats usually have one or a few animals in one enclosure. The Fish and Pet table are connected to the Transaction table. This is a many to many relationship because many people can buy fish and multiple transactions are part of the same database. This applies to pets and transactions as well. Aquarium and Fish are connected and it is a one to one relationship because there is one type of fish in every aquarium. Enclosure and Pets are connected and it is a one to one relationship because there is one type of pet for every enclosure. Since every fish and pet need food, it is a many to many relationship between Food and Fish and Food and Pet. Each fish and pet consumes food. Each pet and fish are stored in an enclosure or aquarium, respectively. Each customer buys a fish or pet and is recorded in the transaction table. Food is then connected to a separate FoodPurchase table because the people who are on the loyalty program get a discount on food for pets and fish. LoyaltyID is a secondary key because it is different than a normal ID, where the customer does not get a discount.
